```python

def shortTerm(current_date):
    tickers =ticker_constant
    ticker_points = {}

    for t in tickers:
        stock_data = databack(ticker=t+'.NS',end_date=current_date,start_date=current_date-timedelta(days=365))
        points = bb(stock_data) + WMA(stock_data) + adx(stock_data)
        ticker_points[t] = points

    best_ticker = max(ticker_points, key=ticker_points.get)
    return best_ticker
```

```python
from datetime import *
def backtest_strategy():


    end_date = datetime.now()
    start_date = end_date - timedelta(days=365*2)
    current_date = start_date


    backtest_results = []

    while current_date < end_date:
        best=shortTerm(current_date)
        stock_data = databack(ticker=best +'.NS',end_date=current_date+timedelta(days=14),start_date=current_date)
        closing_prices = stock_data['Close']
        
        max_close = max(closing_prices)
        min_close = min(closing_prices)
        current_close = closing_prices[0]

        percentage_profit = ((max_close - current_close) / current_close) * 100
        percentage_loss = ((current_close - min_close) / current_close) * 100

        backtest_results.append({"Stock":best,"Date":current_date,'Percentage_Profit': percentage_profit, 'Percentage_Loss': percentage_loss})

        current_date += timedelta(days=14)

    return backtest_results

    
```

```python

def midTerm(current_date):
    tickers =ticker_constant
    ticker_points = {}

    for t in tickers:
        stock_data = databack(t + ".NS",end_date=current_date,start_date=current_date-timedelta(days=365))
        points = RSI(stock_data) + adx(stock_data) + OBV(stock_data)
        ticker_points[t] = points

    best_ticker = max(ticker_points, key=ticker_points.get)
    return best_ticker



    
```
